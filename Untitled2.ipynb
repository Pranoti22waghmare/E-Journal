```python
m = 0
b = 0
lr=0.01
x1 = df['x1']
y1 = df['Target']
```

```python
import numpy as np 
def GD(x,y,b,m,lr):
    for xi,yi in zip(x,y):
        dldb = -2 * np.sum(yi-m*xi-b)
        dldm = -2 *np.sum((yi-m*xi-b)*xi)
        b = b-(lr*dldb)
        m = m-(lr*dldm)
    return m,b
```

```python
for i in range(600):
    m,b = GD(x1,y1,b,m,lr)
    print(m,b)
```
